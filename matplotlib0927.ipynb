##matplotlibグラフの表示ライブラリ


###ライブラリの読み込み

```python
#読み込み表示
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

##グラフ出力するデータの準備

```python
t=np.arange(0.0,2.0,0.01)#0.0～2.0まで0.0刻み
#sinカーブを描くために、元データのsin値を取得して変数sに設定
s=1+np.sin(2*np.pi*t)

#グラフの描画
fig,ax=plt.subplots()
ax.plot(t,s)
#グラフaxラベルを設定する
ax.set(xlabel='time(s)',ylabel='voktage(mv)',
    title='About as simple asit gets, folks')
#グラフaxにグリッドを付ける
ax.grid()

#pyplotで表示
#ax
plt.show()
```
